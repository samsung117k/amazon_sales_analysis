# ipynb notebook for simple data cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv("amazon.csv")
```

```python
df.replace("₹", "", regex=True, inplace=True)
df["rating_count"].replace(",", "", regex=True, inplace=True)
df["actual_price"].replace(",", "", regex=True, inplace=True)
df["discounted_price"].replace(",", "", regex=True, inplace=True)
```

```python
df = df.dropna()
```
